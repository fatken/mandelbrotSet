# Mandelbrot set
Obtained from the quadratic recurrence equation
$$Z_{n+1} = Z_n^2 + C$$
with $z_0 = C$, where points C in the complex plane for which the orbit of $Z_n$ does not tend to infinity are in the set. Setting $Z_0$ equal to any point in the set that is not a periodic point gives the same result.  
## implementation

The coordinates are:  
$coordX, coordY$   
### or  
$coordX + coordY *1j$  
#### C
$C = coordX + coordY * 1j$

So:
$$Z_0 = C = coordX + coordY * 1j$$
$$Z_1 = Z_0**2 + C$$
$$Z_2 = Z_1**2 + C$$
$$...$$
#### Evaluate whether the points are smaller than 2(within the boundary):
$$abs({Z_n}) \leq 2$$

For each iteration, evaluate whether the points are still within the boundary, then assign color to points.  
The points of mandelbrot set are colored according to the number of setps required to reach $r_{max} = 2$

# iterative mandelbrot set calculation

```python
def mandelbrotIterative(z,c):
        return z ** 2 + c
```

```python
#matrix
minX = -3
maxX = 2
minY = -2.5
maxY = 2.5
xrange = np.arange(minX, maxX, 0.0005)
yrange = np.arange(maxY, minY, -0.0005)
xpixel = xrange.shape[0]
ypixel = yrange.shape[0]
coordX, coordY = np.meshgrid(xrange, yrange)
#print(xpixel)
#print(ypixel)
coord = coordX + coordY * 1j
```

```python
iteration = 50
res = np.ones_like(coord, dtype=float)
iterCoord = 0

t1 = time.time()

for i in range(iteration):
    titer1 = time.time()
    color = 1- i/iteration
    if i == 0:
        iterCoord = mandelbrotIterative(coord, coord)
    else:
        iterCoord = mandelbrotIterative(iterCoord, coord)    
    res[abs(iterCoord) <= 2] = color
    titer2 = time.time()
    print('iteration ' + str(i) + ": " + str(titer2 - titer1), end = '\r',)
t2 = time.time()
print('\ntotal time: ' + str(t2 - t1))
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(res, cmap='Greys')#, cmap = 'gray'
plt.imsave('mandelbrotIterativeIteration' + str(iteration) + 'res' +str(xpixel) + 'X' + str(ypixel) + 'Greys' + '.png', res, cmap = 'Greys')#
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(res, cmap='gray')#, cmap = 'gray'
plt.imsave('mandelbrotIterativeIteration' + str(iteration) + 'res' +str(xpixel) + 'X' + str(ypixel) + 'gray' + '.png', res, cmap = 'gray')#
```

## mandelbrot function and experiments

```python
def mandelbrotFun(coord, iteration = 10, boundary = 2):
    iteration = 50
    res = np.ones_like(coord, dtype=float)
    iterCoord = 0

    t1 = time.time()

    for i in range(iteration):
        titer1 = time.time()
        color = 1- i/iteration
        if i == 0:
            iterCoord = mandelbrotIterative(coord, coord)
        else:
            iterCoord = mandelbrotIterative(iterCoord, coord)    
        res[abs(iterCoord) <= boundary] = color
        titer2 = time.time()
        print('iteration ' + str(i) + ": " + str(titer2 - titer1), end = '\r',)
    t2 = time.time()
    print('\ntotal time: ' + str(t2 - t1))
    return(res)
```

## Different boundaries

```python
#matrix
xrange = np.arange(-3, 2, 0.005)
yrange = np.arange(2.5, -2.5, -0.005)
xpixel = xrange.shape
ypixel = yrange.shape
coordX, coordY = np.meshgrid(xrange, yrange)
coord = coordX + coordY * 1j

boundaries = [-1, 0, 0.1, 0.2, 0.5, 1, 2, 100]
res = []

fig = plt.figure(figsize=(20, 10))
fig.suptitle('mandelbrot with different boundaries', fontsize=20)
for i in range(len(boundaries)):
    res = mandelbrotFun(coord,50, boundary=boundaries[i])
    plt.subplot(2, 4, i+1)
    plt.title("boundary: " + str(boundaries[i]))
    plt.imshow(res, cmap = 'jet')
    
fig.tight_layout()
fig.subplots_adjust(top=0.88)
```

## Different matrices

```python
#matrix
xrange = np.arange(-0.6, -0.4, 0.00004)
yrange = np.arange(-0.1, -0.3, -0.00004)
xpixel = xrange.shape
ypixel = yrange.shape
coordX, coordY = np.meshgrid(xrange, yrange)
#print(xpixel)
#print(ypixel)
coord = coordX*1j + coordY *2j

res = mandelbrotFun(coord,50, boundary=2)
```
